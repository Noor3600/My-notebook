```python
# Import tools to play with data and make pretty pictures
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

# Load flower data
iris = load_iris()
data = pd.DataFrame(data=iris.data, columns=iris.feature_names)
data['species'] = iris.target
data['species'] = data['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})

# Show the first 5 flowers
print(data.head())
```

```python
# Check if thereâ€™s anything missing
print("Missing data check:")
print(data.isnull().sum())

# Learn about the data
print("\nData Info:")
print(data.info())
```

```python
# Average size of flowers by species
print("\nAverage Flower Measurements by Species:")
print(data.groupby('species').mean())
```

```python
# Line chart for average sepal width
species_mean = data.groupby('species')['sepal width (cm)'].mean()
species_mean.plot(kind='line', marker='o', title='Average Sepal Width by Species')
plt.ylabel('Sepal Width (cm)')
plt.grid()
plt.show()
```

```python
# Bar chart for average petal length
species_mean_petal = data.groupby('species')['petal length (cm)'].mean()
species_mean_petal.plot(kind='bar', title='Average Petal Length by Species')
plt.ylabel('Petal Length (cm)')
plt.xticks(rotation=0)
plt.show()
```

```python
# Histogram for sepal length
sns.histplot(data['sepal length (cm)'], kde=True)
plt.title('Sepal Length Distribution')
plt.xlabel('Sepal Length (cm)')
plt.show()
```

```python
# Scatter plot for sepal length and width
sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', hue='species', data=data)
plt.title('Sepal Length vs Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend(title='Species')
plt.show()
```
